• Un jeu de données sur les passagers pour lequel le champ 'Survived' (qui indique la survie

Le but du jeu est de construire une fonction qui prenne en entrée les informations sur un

données.

4. Y-a-t-il des données manquantes ? Sur quelles propriétés ? 

```python
df.isnull().sum()
```

```python
new_df = df[["Age" , "Pclass" , "Sex" , "Survived"]]
```

```python
new_df.head(3)
```

5. Ajouter un champ au DataFrame pour indiquer l'existence ou non d'une donnée
manquante sur cette propriété.

```python
df['age_Null'] = df["Age"].isna()
df.head(5)
```

6. Remplacer les données manquantes par des valeurs spéciales. (Utiliser la fonction mean)

```python
# replacing na values in NAN with mean
new_df["Age"] = new_df["Age"].fillna(np.mean(df.Age))
```

```python
new_df.head(4)
```

```python
new_df.isnull().sum()
```

7. Ajuster les variables catégorielles en des variables numérique.

```python
new_df = new_df.replace({
    "Sex":{"male" : 1 , "female" : 0}
})
```

```python
new_df.head(2)
```

8. Construire le Modèle d'arbre de décision

```python
X = new_df.iloc[ : , : -1]
Y = new_df.iloc[ : , -1]
```

```python
model = tree.DecisionTreeClassifier(max_depth=3)
DcM = model.fit(X , Y)
tree.plot_tree(DcM)
plt.show()
```

9. Diviser le jeu de données en deux partie (Training, Testing)

```python
x_train , x_test , y_train , y_test = train_test_split(X , Y ,  train_size=0.75)
```

12. Calculer la prédiction
13. Calculer le score de Training et de Testing

```python
prediction = DcM.predict(x_test)
```

```python
accuracy_score(y_test , prediction)
```

14. Appliquer l’algorithme de la régression Logistique sur le même jeu de données.


```python
modLog = LogisticRegression()
modLog
```

15. Appliquer l’apprentissage


```python
modLog.fit(x_train , y_train)
```

16. Calculer le score de Training et de Testing


```python
accuracy_score(y_test , modLog.predict(x_test))
```
