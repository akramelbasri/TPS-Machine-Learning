### Etape 1 : regulariser les phrases de df :

```python
phrase1 = df.values[0 ,0]
propre_ph = re.sub("[^a-zA-Z]" , " " , phrase1)
# propre_ph
print(f'- La phrase initial: {phrase1.lower()} \n- Phrase regulariser :{propre_ph.lower()}')
```

```python
listWords = propre_ph.split()
for i in listWords:
    print(i)
```

- afficher la racine d'un certain mot 

```python
nltk.stem.porter.PorterStemmer().stem('speaking')
```

- recine de mot (l'origine) (mehode stem)

```python
from nltk.stem.porter import PorterStemmer
racine = PorterStemmer()
# supprimer les stopwords
Non_StpWds = []
words_racines = []

for word in listWords:
    if(word not in set(stopwords.words('english'))) : 
        Non_StpWds.append(word)
        words_racines.append(racine.stem(word))
print(Non_StpWds)
print(words_racines)
_corpus = " ".join(words_racines)
_corpus
```

```python
print(f'Les entités : {listWords} \nSuppression des stopwords: {Non_StpWds} \nOrigine des mots : {words_racines} \nCorpus : {_corpus}' )
```

- parcourire de tout les phrase de notre dataset :

```python
phrase_propre = []
for i in range(len(df)) : 
    phrase = re.sub("[^a-zA-Z]" , " " , df.values[i , 0]) # Eliminer les carac
    phrase = phrase.lower() # transformer en lower
    phrase = phrase.split() # resemblage
    ps = PorterStemmer() # returne l'origine d'un mot
    phrase = [ps.stem(mot) for mot in phrase if(mot not in set(stopwords.words('english'))) ]
    phrase = " ".join(phrase)
    phrase_propre.append(phrase)
```

```python
print(phrase_propre)
```

```python
print(df)
```

# L'encodage de donnees ,et l'application d'un algorithme de machine learning

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
cv
```

```python
matrice = cv.fit_transform(phrase_propre) # sparse matrice : binaire 0/1
print(matrice.shape)
```

```python
X = matrice.toarray()
Y = df["Liked"]
Y
```

```python
cv.get_feature_names()
```

- creation de notre dataframe

```python
# cv.get_feature_names()

x_df = pd.DataFrame(data = X , columns=cv.get_feature_names())
y_df = pd.DataFrame(data= Y.values , columns=["Target"])
dataFrame = pd.concat([x_df , y_df] , axis=1)
```

```python
dataFrame # propre dataframe (applique d'un algorithme de machine learning) 
```

 - pour bien visualiser les coordonnees de la matrice en faire le Transposer

```python
dataFrame.T 
```

- pipeline de machine learning

```python
x = dataFrame.iloc[ : , 0 : -1 ]
y = dataFrame.iloc[ : , -1 ]
```

- creer le Training set et le Test set :

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(x , y , train_size=0.8 , random_state=1)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

# Application de l'lgorithme Random Forest:

- importation :

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
#n_jobs=2) # accelirer la vitesse de l'entrainement
```

Entrainement de modele : 

```python
%timeit model.fit(X_train , Y_train) # Le tempe de l'entrainment
```

```python
model.fit(X_train , Y_train)
```

La prédictions

```python
y_predicts_binaire = model.predict(X_test)
y_predicts_proba = model.predict_proba(X_test)
```

Evaluation de modele

```python
# score de modele
score_model = model.score(X_train , Y_train)
score_model
```

```python
# score de prediction
score = model.score(X_test , Y_test)
score
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
# La matrice de confusion
mat = confusion_matrix(y_predicts_binaire , Y_test)
mat
```
